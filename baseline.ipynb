# Получение необходимой информации по адресу/координатам (сваггер)
https://geo.hack-it.yazzh.ru/
https://geo.hack-it.yazzh.ru/redoc

```python
import requests
from pprint3x import pprint
response = requests.get("https://geo.hack-it.yazzh.ru/api/v2/geo/buildings/search/", params={"query":"Аптекарский проспект 18"}).json()
pprint(response)
```

по адресу получили идентификатор дома (id) 209676, а также его координаты latitude': 59.9156319, 'longitude': 30.2985272
получаем информацию по дому

```python
response = requests.get("https://geo.hack-it.yazzh.ru/api/v2/geo/buildings/209676").json()
pprint(response)
```

по идентификатору дома получили наименование района (district): 'name': 'Петроградский'
делаем запрос по Афише событий: обратите на формат даты.
Для всех методов с swagger есть подробное описание query params

```python

response = requests.get("https://hack-it.yazzh.ru/afisha/all/?start_date=2024-11-25T09%3A24%3A43&end_date=2025-11-25T09%3A24%3A43&location_latitude=59.9156319&location_longitude=30.2985272&location_radius=1&page=1&count=10").json()
pprint(response)
```
